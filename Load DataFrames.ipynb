```python
# fillna

# Replace NaNs with blanks if the column is a string, so we don’t get
# objects in the column
def fillna_col(series):
    if series.dtype is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
# TODO: Define nullify here

# Replace \Ns with NaNs
def nullify(x):
    if x == "\\N":
        return float('NaN')  # TODO: replace this!
    else:
        return x
```

```python
# TODO: regularize and remove nulls according to Step 2
def clean(df):
    return df.applymap(nullify).apply(fillna_col)
airports_df = clean(airports_df)
airlines_df = clean(airlines_df)
flights_df = clean(flights_df)
routes_df = routes_df.applymap(nullify)
routes_df = routes_df.dropna(axis=0, how='any', subset=['r_airline_id', 'source_airport_id', 'destin_airport_id'])
routes_df = routes_df.apply(fillna_col)

airlines_df
```

```python
# TODO: change column types according to Step 2
routes_df['r_airline_id'] = routes_df['r_airline_id'].astype(int)
routes_df['source_airport_id'] = routes_df['source_airport_id'].astype(int)
routes_df['destin_airport_id'] = routes_df['destin_airport_id'].astype(int)
```

```python
# TODO: read and clean raw text file and store in clean_incidents
with open("incidents_raw.txt","r") as incident:
    read_incident = incident.read()
    replace_incident = read_incident.replace("[edit]", "")
    list_incident = replace_incident.split("/n")

ind = list_incident.index("1997")
list_incident = list_incident[ind:]

new_list =[]
for i in list_incident:
    if len(i) == 4:
        flag = i
    else:
        new_list.append(flag + " " +i)

clean_list = []
for i in new_list:
    if "flight" or "-" in new_list:
        print(i)
        clean_list.append(i)
```

```python
# TODO: convert clean_incidents into dataframe incidents_df
incidents_df = pd.DataFrame(columns=['Date', 'Airline', 'FlightNum'])

import re
year =[]
month =[]
date =[]
airl =[]
fli_n = []
dict_month = {'January':'1', 'February':'2','March':'3','April':'4','May':'5', 'June':'6','July':'7','August':'8',
              'September':'9','October':'10','November':'11','December':'12'}
for i in clean_list:
    m = re.search('(\d{4})\s(\w+)\s(\d{1,2})\s[–].*?\s([^,:]+)\sFlight\s(\d+)\W', i)
    if m:
        n = m.group(1)
        a = m.group(2)
        b = m.group(3)
        c = m.group(4)
        d = m.group(5)
        year.append(n)
        month.append(a)
        date.append(b)
        airl.append(c)
        fli_n.append(d)
month_new = []        
for i in month:
    month_new.append(dict_month.get(i))

time = []
for i in range(len(year)):
    j = year[i]+'-'+month[i]+'-'+date[i]
    time.append(j)

s_time = pd.Series(time)
s_airl = pd.Series(airl)
s_fli_n = pd.Series(fli_n)
incidents_df['Date'] = s_time.values
incidents_df['Airline'] = s_airl.values
incidents_df['FlightNum'] = s_fli_n.values
incidents_df.drop_duplicates() 
print(len(incidents_df))
```

```python
# TODO: change column type of 'Date'
incidents_df['Date'] = pd.to_datetime(incidents_df['Date'])
print(incidents_df.head(100))
```
